## 向量/矩阵的初始化
  

```python
import numpy as np
print(f'这是一个数组初始化{np.array([1,2])}')
print(f'这是一个行向量初始化{np.array([[1,2]])}')
print(f'这是一个列向量初始化{np.array([[1],[2]])}')
print('总结：列向量和行向量是一个特殊矩阵，所以矩阵初始化也要一个二维数组')
```

```python
import numpy as np
a = np.array([[1],[2]])
b = np.array([[4],[3],[2]])
print(a@b)
```
