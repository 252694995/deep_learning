# 知识点总览图


## 复数
  复数包含了虚数和实数
### 实数
实数的记号为R，实数包含了无理数和有理数。实数集合可以用数轴来展示。  
  

#### 有理数
有理数的记号为Q，包含了整数、分数、小数（有限小数和无限循环小数），可以用两个整数的商来表示：a/b.

##### 整数的性质
- 整数相加、相减、相乘的结果还是整数
- 奇偶性：被2整除是偶数，否则是奇数

#### 无理数
数轴上除了有理数之外的，都是无理数，也可以说不能用两个整数的商表示的。

## 向量
若干数字拍成一行或一列，并且用中括号括起来，得到的数组叫做向量（vector），需注意的是，向量的维度是1（维度就是形状1xD或nX1）
## 矩阵
矩阵可以看做是一组列向量从左到右叠加，或一组行向量从上到下叠加出来的。  
  
### 矩阵转置
矩阵转置指的是将矩阵列行互换得到的新矩阵  
  
从以上图来看，把矩阵看成一组列向量，然后把列向量全部转置成行向量  
### 矩阵形状
1、行数和列数相同的矩阵叫方阵  
2、列向量的元素全部为1叫全1向量
3、列向量的元素全部维0叫零向量
  
### 矩阵加减
前提：矩阵加减要求两个矩阵的形状相同。
计算过程：对应位置进行加减
目的：批量加减
