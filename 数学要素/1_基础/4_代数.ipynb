# 知识总览图
  

##  集合
### 集合与元素关系
集合与元素之间的关系只有属于和不属于。 

### 集合与集合关系
集合与集合之间有子集、真子集、并集、交集、补集、相对补集。  
#### 补集和相对补集的区别
补集是集合和全集之间的关系，设$\Omega$是一个集合，$A$是$\Omega$的一个子集，补集是在$\Omega$集合中但不在$A$中的集合；  
相对补集是集合和集合的关系，设$A$和$B$是一个集合,元素在$A$中不在$B$中的集合；

## 多项式
算数研究对象是已知数，而代数研究对象是未知数，通常是用表达式来描述，表达式有区分多项式和单项式

### 基础
项：在表达式中，由操作符连接的操作数。  
元：变量（代数研究的是未知数，也就是变量）  
系数：在代数表达式中与变量的乘积的常数因子   
  
例如：P(x)=$ax^2+x+1$  
项：三项，分别是$ax^2、x、1$  
元：一元，是x  
系数：$ax^2的系数是a，x的系数是1，1由于没有存在变量，所以没有系数$

## 函数
一个定集合X，对X中的元素施加映射法则f,记作函数f(x),得到的结果y=f(x)属于集合Y（对于函数x，y只有唯一的值与之对应，也就是一一映射或多对一映射关系）。  
x也被称为输入值或定义域，y被称为输出值或值域，x是自变量，y是因变量。
