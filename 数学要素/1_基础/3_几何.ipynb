# 知识总览图
  


## 几何体
### 点
点确定空间的一个位置，点本身没有长度、面积等几何属性。

### 线
- 直线： 任意两点可以画一条直线，直线可以沿两个方向无限延伸，没有端点
- 线段：线段是有两个端点

### 几何变换
  
#### 平移
将图形沿着定的方向和距离移动。平移不改变图形的形状、大小、方向，只改变了位置。
#### 旋转
将图形围绕着一个定点旋转一定的角度。旋转不改变形状和大小，只改变了位置和方向。
#### 镜像
镜像是将图形沿着一个给定的线进行对称翻转的操作。
#### 缩放
#### 投影

## 角度和弧度
### 弧度
定义为的圆弧长除以半径，作用是角度的无单位表达。
- 角度约等于57.2958对应的圆弧长除以半径，商就是1弧度，简写作1rad。
2π = 360度推理过程：
因为圆一周360度，周长是2πr/r=360度

### 余弦定理
余弦定义成夹角所对边的邻边除以斜边
  


## 向量
向量是有方向和大小的，例如从家到学校表达。向量可以随便平移，不影响方向和大小。
### 向量加法  
共起点、尾首相连结果的向量都是起点到终点。
### 向量减法  
共起点:oa-ob=oa+(-ob)=ba
### 向量数乘 
延长或缩短
### 向量内积  
ab=|a||b|cos
相当于一个向量在另一个向量的投影长度
