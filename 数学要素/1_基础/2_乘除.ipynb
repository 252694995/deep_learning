## 向量乘法
### 标量乘法
一个系数乘以一个向量，结果还是向量。  
  

### 向量内积
向量内积又叫标量积，两个向量相乘，结果是一个标量。
向量内积运算规则：两个形状一样的向量，对应位置元素一一相乘后再求和  
  

## 向量逐项积
与向量内积相同点是相同形状向量相乘，区别是不同的计算规则。
运算规则：两个形状相同的向量相乘，对应元素进行相乘，得到一个形状相同的向量 ,一般运用在算术的批量运算上。  

## 向量转置乘法
向量转置乘法遵从矩阵乘法一个规则，就是第一个向量mxp和第二个向量pxn乘法，要求左向量列等于右向量行，得到一个mxn形状的向量。 被乘数和乘数都是列向量。
  

### 被乘数转置
两个等长的列向量，第一个向量转置乘以第二个向量，结果是标量，等价于两个向量的标量积。
  

### 乘数转置
两个等长的列向量，列向量乘以第二个向量（转置），结果为方阵。

## 矩阵乘法
运算规则：A和B两个矩阵相乘，生成一个A行B列的矩阵向量，矩阵A的第一行元素与B向量的  
第一列对应元素相乘相加，计算结果放在第一行一列。  元素不一定是标量，也可以是向量。

## 矩阵除法
除法的意义是根据上下文，在数学上，除法是乘法的逆运算，是还原操作数的运算。
在矩阵上，矩阵B除以矩阵A，实际上是将矩阵A转化逆矩阵A-1，然后计算标量集。
